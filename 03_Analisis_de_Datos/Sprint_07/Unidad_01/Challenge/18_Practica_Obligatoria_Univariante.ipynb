### # 2.1

````python
def card_tipo(df, umbral_categoria = 9, umbral_continua = 30):
    df_temp = pd.DataFrame([df.nunique(), df.nunique() / len(df) * 100, df.dtypes]).T
    df_temp = df_temp.rename(columns = {0: 'Card', 1: '%_Card', 2: 'Tipo'})
    
    df_temp['Tipo_sugerido'] = 'Categórica'
    df_temp.loc[df_temp['Card'] == 1, '%_Card'] = 0.00
    df_temp.loc[df_temp['Card'] == 2, 'Tipo_sugerido'] = 'Binaria'
    df_temp.loc[df_temp['%_Card'] >= umbral_categoria, 'Tipo_sugerido'] ='Numérica discreta'
    df_temp.loc[df_temp['%_Card'] >= umbral_continua, 'Tipo_sugerido'] = 'Numérica continua'
    
    return df_temp
````

````python
card_tipo(df_titanic)
````

````python
types = card_tipo(df_titanic).reset_index()
categoric = [types['index'].iloc[i] for i, e in enumerate(types['Tipo_sugerido']) if e == 'Categórica']
categoric
````

````python
numeric = [types['index'].iloc[i] for i, e in enumerate(types['Tipo_sugerido']) if (e == 'Numérica discreta') or (e == 'Discreta contínua')]
numeric
````

````python
def get_CV(df, column):
    column = [column] if type(column) == str else column
    return df[column].describe().T['std'] / df[column].describe().T['mean'] * 100
````
