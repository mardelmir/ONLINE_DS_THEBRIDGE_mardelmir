Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:

#### 1. Ciudades más pobladas
Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
# Preparación del DataFrame
df_pop = df.loc[df['population_total'].nlargest(5).index]
df_pop['population_total'] = df_pop['population_total'] / 1000
add_space = lambda text: re.sub(r'(?<!^)(?=[A-Z])', ' ', text)

for city in df_pop['city'].to_list():
    df_pop['city'] = df_pop['city'].str.replace(city, add_space(city))

# Creación del diagrama
colors = ['blue', 'gray', 'r', 'olive', 'brown']
plt.bar(df_pop['city'], df_pop['population_total'], color = colors, alpha = 0.5)

for i in df_pop['city'].index:
    population = df_pop.loc[i,'population_total']
    city = df_pop.loc[i,'city']
    plt.text(city, population + 100, f'{population}', ha = 'center')

plt.title('Cuidades más pobladas de California')
plt.xlabel('Cuidades')
plt.ylabel('Habitantes (miles)')
plt.ylim(0, 5000)
plt.show()
```

#### 2. Histograma
Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
font = {'fontname': 'Bradley Hand itc', 'fontsize': 18}
plt.xlabel('Superficie ($km^2$) ', **font)
plt.ylabel('Número de Ciudades', **font)

patch = mpatches.Patch(color = 'steelblue', label = 'Hist Area')
plt.legend(loc = 'upper right', handles = [patch])
plt.show()
```

#### 3. Scatter plot
El scatter de área de agua de un ciudad con su poblacion.

```python
plt.scatter(df['area_water_km2'], df['population_total'] / 1000, color = 'green', alpha = 0.7)
plt.text(65, 3500, 'Outliers', ha = 'center', va = 'center')
outliers = (df['area_water_km2'] > 80) & (df['population_total'] > 1000)

x = df['area_water_km2'][outliers].values
y = (df['population_total'][outliers] / 1000).values

plt.annotate('', xy = (x[0], y[0]), xytext = (70, 3600), arrowprops = {'arrowstyle': '->', 'connectionstyle': 'arc3,rad=0'})
plt.annotate('', xy = (x[1], y[1]), xytext = (70, 3400), arrowprops = {'arrowstyle': '->', 'connectionstyle': 'arc3,rad=0'})

plt.xlabel('Área de agua en la ciudad (km²)')
plt.ylabel('Población (miles de habitantes)')
plt.title('Población por kilómetro cuadrado de agua')
plt.show()
```

#### 4. Box plot
El boxplot de la altitud de las ciudades.

```python
# 4. Boxplot
# plt.boxplot(df['elevation_m'], vert = False, patch_artist = True, boxprops = {'facecolor':'red'}) # Originial

plt.boxplot(df['elevation_m'], vert = False, patch_artist = True, boxprops = {'facecolor':'red'}, medianprops = {'color': 'black'})
plt.title('Diagrama de caja de la Altitud de las ciudades de California');
plt.xlabel('Altitud (m)')
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (20, 20))

# 1. Diagrama de barras
df_pop = df.loc[df['population_total'].nlargest(5).index]
df_pop['population_total'] = df_pop['population_total'] / 1000
add_space = lambda text: re.sub(r'(?<!^)(?=[A-Z])', ' ', text)

for city in df_pop['city'].to_list():
    df_pop['city'] = df_pop['city'].str.replace(city, add_space(city))

colors = ['blue', 'gray', 'r', 'olive', 'brown']
ax[0,0].bar(df_pop['city'], df_pop['population_total'], color = colors, alpha = 0.5)

for i in df_pop['city'].index:
    population = df_pop.loc[i,'population_total']
    city = df_pop.loc[i,'city']
    ax[0,0].text(city, population + 100, f'{population}', ha = 'center')

ax[0,0].set_title('Cuidades más pobladas de California')
ax[0,0].set_xlabel('Cuidades')
ax[0,0].set_ylabel('Habitantes (miles)')
ax[0,0].set_ylim(0, 5000)


# 2. Histograma
ax[0,1].hist(df['area_total_km2'], bins = 50, color = 'steelblue', alpha = 0.7)

patch = mpatches.Patch(color = 'steelblue', label = 'Hist Area')
ax[0,1].legend(loc = 'upper right', handles = [patch]);

font = {'fontname': 'Bradley Hand itc'}
ax[0,1].set_xlabel('Superficie ($km^2$) ', **font, fontsize = 18)
ax[0,1].set_ylabel('Número de Ciudades', **font, fontsize = 18)


# 3. Scatter plot
ax[1,0].scatter(df['area_water_km2'], df['population_total'] / 1000, color = 'green', alpha = 0.7)
ax[1,0].text(65, 3500, 'Outliers', ha = 'center', va = 'center')
outliers = (df['area_water_km2'] > 80) & (df['population_total'] > 1000)

x = df['area_water_km2'][outliers].values
y = (df['population_total'][outliers] / 1000).values

ax[1,0].annotate('', xy = (x[0], y[0]), xytext = (69, 3550), arrowprops = {'arrowstyle': '->', 'connectionstyle': 'arc3,rad=0'})
ax[1,0].annotate('', xy = (x[1], y[1]), xytext = (67, 3450), arrowprops = {'arrowstyle': '->', 'connectionstyle': 'arc3,rad=0'});

ax[1,0].set_title('Población por kilómetro cuadrado de agua')
ax[1,0].set_xlabel('Área de agua en la ciudad (km²)')
ax[1,0].set_ylabel('Población (miles de habitantes)')

# 4. Boxplot
ax[1,1].boxplot(df['elevation_m'], vert = False, patch_artist = True, boxprops = {'facecolor':'red'}, medianprops = {'color': 'black'})
ax[1,1].set_title('Diagrama de caja de la Altitud de las ciudades de California');
ax[1,1].set_xlabel('Altitud (m)')
plt.show()
```

```python
mas_pobladas = df.nlargest(5, 'population_total')
cities = mas_pobladas['city']
pops = mas_pobladas['population_total']
colores = ['blue','gray','coral','olive','brown']
plt.figure(figsize=(7,8))
bars = plt.bar(cities, pops/1000, color=colores,alpha=0.4)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.1f}', ha='center',va='bottom')
plt.ylim(0,5000)
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.title("Ciudades mas pobladas de California")
```

```python
def separa_nombre(nombre):
    nuevo = nombre[0]
    for letra in nombre[1:]:
        if letra == letra.upper():
            nuevo += ' '
        nuevo += letra
    return nuevo

ciudades_mas_pobladas = df.loc[df.population_total.nlargest(5).index, ['city', 'population_total']]
ciudades = [separa_nombre(ciudad) for ciudad in ciudades_mas_pobladas['city'].to_list()]
colores = ['blue', 'gray', 'red', 'olive', 'brown']
plt.figure(figsize = (10,5))
plt.bar(x = ciudades, height = ciudades_mas_pobladas['population_total'] / 1000, color = colores, alpha = 0.5)
plt.ylim([0, 5000])
plt.ylabel('')
plt.xlabel('Ciudades')
```
